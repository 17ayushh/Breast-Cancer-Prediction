```python
import numpy as np
import pandas as pd
import sklearn.datasets 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
df = pd.DataFrame(breast_cancer_dataset.data ,columns = breast_cancer_dataset.feature_names)
```

```python
df.head()
```

```python
df['label'] = breast_cancer_dataset.target
```

```python
X=df.drop(columns='label',axis=1)
y=df['label']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

## Building the Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid') # while deciding the output layer remember , no of neuron depicts the classes as it is binary i am using 2 class

    
])
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train_std,y_train,epochs=10,validation_split=0.1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.legend(['training data','validation data'],loc = 'lower right')
```

```python
loss , accuracy = model.evaluate(X_test_std,y_test)
```
